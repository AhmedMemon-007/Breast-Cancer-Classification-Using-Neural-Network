```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Loading the data from sklearn
breast_cancer_data=sklearn.datasets.load_breast_cancer()
```

```python
#Loading the data to a DataFrame
cancer_data=pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
```

```python
#Adding the target column to the dataframe
cancer_data['label']=breast_cancer_data.target
```

*Seperating features and targets*

```python
x= cancer_data.drop(['label'],axis=1)
y=cancer_data['label']
```

```python
print(x)
```

```python
print(y)
```

```python
#Splitting into Test and Test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print('Training set size: ',x_train.shape)
print('Test set size: ',x_test.shape)
```

*Standardised the data*

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
print(x_train_std)
```

#Building a Neural Network

```python
#Import TensorFlow and Keras
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3)
#The line tf.random.set_seed(3) sets a fixed seed for TensorFlows random number generator.
#This ensures that any random operations (like initializing weights in a neural network) produce the same results every time you run
#the code. Setting a seed helps make results reproducible, so you get consistent outputs, which is useful for debugging and comparing
#model performance across runs.
```

```python
#setting up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),       #Input layer   30 --> features in dataset
    keras.layers.Dense(20,activation='relu'),     #1 Hidden layer 20 = e.g. we want 20 neurons activation function = relu =rectified linear units
    keras.layers.Dense(2,activation='sigmoid')   # Output layer  2 = output neuron activation= sigmoid --> is just like the logistic regression function

    #We have 2 neurons in the output layer because we have only 2 different classes in target column 0 and 1
    ])
```

```python
#Compiling the neural network

model.compile(optimizer='adam',
               loss= 'sparse_categorical_crossentropy',
               metrics= ['accuracy'])
```

*important note= <Br>
if there is output as :<br> dog = 1  cat = 2  horse = 3<br>
then loss = sparse_categorical_crossentropy*
<Br> if there is one-hot-encoding :<br>
cat = [1,0,0]  dog = [0,1,0]  horse = [0,0,1]<br>
then use loss = categorical_crossentropy

```python
#Training the neural network

history = model.fit(x_train_std, y_train,validation_split=0.1,epochs=10)
```

*visualizing Accuracy and Loss*

```python
# Plotting Model Accuracy
plt.plot(history.history['accuracy'])         # Training accuracy
plt.plot(history.history['val_accuracy'])     # Validation accuracy

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc='lower right')
plt.show()
```

```python
# Plotting Model Accuracy
plt.plot(history.history['loss'])         # Training accuracy
plt.plot(history.history['val_loss'])     # Validation accuracy

plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc='upper right')
plt.show()
```

*Accuracy of the model on the test data*

```python
loss,accuracy= model.evaluate(x_test_std,y_test)
```
